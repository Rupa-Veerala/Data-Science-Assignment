```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from fpdf import FPDF
```

Step 2: Load the Datasets

```python
# Load datasets
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
transactions = pd.read_csv("Transactions.csv")
```

Step 3: Inspect the Data

```python
#Check for missing values, data types, and basic statistics.
print("Missing values in Customers:")
print(customers.isnull().sum())

print("\nMissing values in Products:")
print(products.isnull().sum())

print("\nMissing values in Transactions:")
print(transactions.isnull().sum())
```

```python
#Check the data types of each column
print("Data types in Customers:")
print(customers.dtypes)

print("\nData types in Products:")
print(products.dtypes)

print("\nData types in Transactions:")
print(transactions.dtypes)
```

```python
#Get the number of rows and columns, and preview the first few rows
# Structure and preview
print("Structure of Customers:")
print(customers.shape)
print(customers.head())

print("\nStructure of Products:")
print(products.shape)
print(products.head())

print("\nStructure of Transactions:")
print(transactions.shape)
print(transactions.head())
```

```python
#Summary Statistics: Get an overview of numeric columns.
print(customers.describe())
print(products.describe())
print(transactions.describe())
```

```python
#Column Names: Print all column names to understand the dataset structure.
print(customers.columns)
print(products.columns)
print(transactions.columns)
```

Step 4. Analyze Customer Distribution by Region

```python
#Check how customers are distributed across different regions.
import matplotlib.pyplot as plt
import seaborn as sns

# Count customers by region
region_distribution = customers['Region'].value_counts()

# Plot the distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=region_distribution.index, y=region_distribution.values, palette="viridis")
plt.title("Customer Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.show()
```

Step 5. Analyze Product Categories and Their Sales

```python
#Calculate sales by product category.
# Merge Transactions and Products datasets
merged_data = transactions.merge(products, on='ProductID', how='left')

# Group by Category and calculate total sales using 'TotalValue'
category_sales = merged_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)

# Plot product category sales
plt.figure(figsize=(8, 5))
category_sales.plot(kind='bar')
plt.title('Total Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales Value')
plt.xticks(rotation=45)
plt.show()
```

Step 6. Look at Trends in Transactions (e.g., by Month/Year)

```python
#Analyze transaction trends over time.
# Convert Date column to datetime
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])

# Extract Year and Month
transactions['YearMonth'] = transactions['TransactionDate'].dt.to_period('M')

# Group by YearMonth and count transactions
transaction_trends = transactions.groupby('YearMonth')['TransactionID'].count()

# Plot transaction trends
plt.figure(figsize=(10, 5))
transaction_trends.plot(kind='line', marker='o', color='teal')
plt.title("Trends in Transactions Over Time")
plt.xlabel("Year-Month")
plt.ylabel("Number of Transactions")
plt.grid()
plt.show()
```

Step 7. Identify Top Customers Based on the Total Value of Purchases

```python
#Find customers who spend the most.
# Group by CustomerID and calculate total purchase value using 'TotalValue'
top_customers = transactions.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(10)

# Plot top customers
plt.figure(figsize=(8, 5))
top_customers.plot(kind='bar')
plt.title('Top 10 Customers by Total Spend')
plt.xlabel('CustomerID')
plt.ylabel('Total Purchase Value')
plt.xticks(rotation=45)
plt.show()
```

Step 8. Find the Most Popular Product Categories

```python
#Identify the categories with the highest number of transactions.
# Group by Category and count transactions
popular_categories = merged_data['Category'].value_counts()

# Plot popular product categories
plt.figure(figsize=(8, 5))
sns.barplot(x=popular_categories.index, y=popular_categories.values, palette="magma")
plt.title("Most Popular Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=45)
plt.show()
```
